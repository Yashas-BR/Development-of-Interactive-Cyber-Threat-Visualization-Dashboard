# 1. Variables in Python

A variable is used to store data in memory.
Python automatically detects the data type.

## Types of Variables
- Integer
- Float
- String
- Boolean


```python
# Integer
age = 21

# Float
height = 5.8

# String
name = "Yashas"

# Boolean
is_student = True

print("Age:", age)
print("Height:", height)
print("Name:", name)
print("Is Student:", is_student)
```

# 2. Data Types in Python

Python supports different data types for different kinds of data.

We can check the type of a variable using the `type()` function.


```python
x = 10
y = 3.14
z = "Hello"
a = False

print(type(x))
print(type(y))
print(type(z))
print(type(a))
```

# 3. Operators in Python

Operators are used to perform operations on variables and values.

## Arithmetic Operators#
+,  - , * , /
- %
- **



```python
a = 10
b = 3

print("Addition:", a + b)
print("Subtraction:", a - b)
print("Multiplication:", a * b)
print("Division:", a / b)
print("Modulus:", a % b)
print("Power:", a ** b)
```

# 4. Conditional Statements in Python

Conditional statements are used to make decisions in a program.

Syntax:
- if
- elif
- else


```python
age = 18

if age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")
```

# 5. For Loop in Python

A for loop is used to iterate over a sequence like list, tuple, or string.


```python
for i in range(1, 6):
    print("Number:", i)
```

# 6. While Loop in Python

A while loop runs as long as the condition is True.


```python
count = 1

while count <= 5:
    print("Count:", count)
    count += 1
```

# 7. Functions in Python

A function is a block of reusable code.

Syntax:
def function_name():


```python
def greet(name):
    return "Hello " + name

print(greet("Yashas"))
```

# 8. Lists in Python

A list is used to store multiple values in a single variable.
Lists are ordered and changeable.


```python
fruits = ["Apple", "Banana", "Mango"]

print("First Fruit:", fruits[0])

fruits.append("Orange")

print("Updated List:", fruits)
```

# 9. Dictionaries in Python

A dictionary stores data in key-value pairs.


```python
student = {
    "name": "Yashas",
    "age": 20,
    "course": "B.Tech"
}

print("Name:", student["name"])
print("Age:", student["age"])
```

# 10. Tuples in Python

A tuple is used to store multiple items in a single variable.
Tuples are ordered but immutable (cannot be changed).


```python
colors = ("Red", "Green", "Blue")

print("First Color:", colors[0])

print("All Colors:", colors)
```

# 11. Sets in Python

A set is a collection of unique values.
Sets do not allow duplicate elements.


```python
numbers = {1, 2, 3, 4, 4, 5}

print("Set Values:", numbers)

numbers.add(6)

print("Updated Set:", numbers)
```

# 12. OOP in Python

OOP is based on the concept of Classes and Objects.

A class is a blueprint.
An object is an instance of a class.


```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print("Name:", self.name)
        print("Age:", self.age)

s1 = Student("Yashas", 20)
s1.display()
```

# 13. Inheritance in Python

Inheritance allows one class to inherit properties from another class.


```python
class Person:
    def greet(self):
        print("Hello!")

class Teacher(Person):
    def subject(self):
        print("Teaching Python")

t1 = Teacher()
t1.greet()
t1.subject()
```

# 14. NumPy Library

NumPy is used for numerical operations and arrays.


```python
import numpy as np

arr = np.array([1, 2, 3, 4, 5])

print("Array:", arr)
print("Mean:", np.mean(arr))
print("Square:", arr ** 2)
```

# 15. Pandas Library

Pandas is used for data analysis and handling datasets.


```python
import pandas as pd

data = {
    "Name": ["Yashas", "Rahul", "Anita"],
    "Marks": [85, 90, 88]
}

df = pd.DataFrame(data)

print(df)
print("Average Marks:", df["Marks"].mean())
```

# 16. Matplotlib - Data Visualization

Matplotlib is used to create graphs and charts in Python.

Common Graphs:
- Line Plot
- Bar Chart
- Scatter Plot


```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y)
plt.title("Line Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()

categories = ["A", "B", "C", "D"]
values = [5, 7, 3, 8]

plt.bar(categories, values)
plt.title("Bar Chart Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [15, 18, 20, 22, 30]

plt.scatter(x, y)
plt.title("Scatter Plot Example")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

# 17. Seaborn - Statistical Data Visualization

Seaborn is built on top of Matplotlib.
It provides more attractive and informative statistical graphics.


```python
!pip install seaborn
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.lineplot(x="total_bill", y="tip", data=tips)

plt.title("Total Bill vs Tip")
plt.show()
```

```python
corr = tips.corr(numeric_only=True)

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
